```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
df = pd.read_csv('dataset - Sheet1.csv')
df['income'] = df['income'].fillna(df['income'].mean())
df['spending_score'] = df['spending_score'].fillna(df['spending_score'].mean())
df['purchase_frequency'] = df['purchase_frequency'].fillna(df['purchase_frequency'].mean())
df['last_purchase_amount'] = df['last_purchase_amount'].fillna(df['last_purchase_amount'].mean())
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])
df['preferred_category'] = df['preferred_category'].fillna(df['preferred_category'].mode()[0])
num_cols = [ 'income', 'spending_score','purchase_frequency','last_purchase_amount']

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df['preferred_category'] = le.fit_transform(df['preferred_category'])
scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
df.to_csv('dataset_cleaned.csv', index=False)
print(df.head())
```
